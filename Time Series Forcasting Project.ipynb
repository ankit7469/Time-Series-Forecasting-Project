```python
# LSTM Time-Series Forecasting Project

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers


# Generate time series -------------------------
np.random.seed(42)
t=np.arange(0,500,0.1)
sine = np.sin(0.04 * t)                  # periodic pattern
trend = 0.002 * t                         # small increasing trend
noise = 0.2 * np.random.randn(len(t))

series=sine+trend+noise

plt.figure(figsize=(10,4))
plt.plot(series)
plt.title("Original Time Series")
plt.show()


#  Convert series into windows
# -----------------------------------------------------
def make_windows(series, window_size, horizon=1):
    X, y = [], []
    for i in range(len(series) - window_size - horizon + 1):
        X.append(series[i:i + window_size])
        y.append(series[i + window_size])
    return np.array(X), np.array(y)
    
WINDOW=30
HORIZON = 1
X, y = make_windows(series, WINDOW, HORIZON)

# reshape: (samples, timesteps, features)
X = X.reshape(X.shape[0], X.shape[1], 1)


#  Train & Test Split
# -------------------------------------------------------
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, shuffle=False)

# Build LSTM model -------------------------------------------

model=keras.Sequential([
    layers.LSTM(64, return_sequences=False, input_shape=(WINDOW, 1)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

# comile model-------------------------------------------

model.compile(optimizer='adam',loss='mse',metrics=['mae'])
model.summary()

# 5) Train model
# ------------------------------------------
EPOCHS = 40
history = model.fit(X_train, y_train,validation_data=(X_val, y_val),epochs=EPOCHS,batch_size=32,verbose=1)

# Plot training curves
plt.figure(figsize=(10,4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title("Training Curves (LSTM)")
plt.show()


#  Predictions on validation set
# --------------------------------------------
val_pred = model.predict(X_val).flatten()

plt.figure(figsize=(12,4))
plt.plot(y_val[:200], label="Actual")
plt.plot(val_pred[:200], label="Predicted")
plt.title("LSTM Predictions vs Actual")
plt.legend()
plt.show()


#  Forecast next future values
# ----------------------------------------------
future_steps = 100
last_window = series[-WINDOW:]
preds = []

current = last_window.copy()

for i in range(future_steps):
    inp = current.reshape(1, WINDOW, 1)
    next_val = model.predict(inp)[0, 0]
    preds.append(next_val)

    # slide window
    current = np.roll(current, -1)
    current[-1] = next_val

# plot future forecast ---------------------
plt.figure(figsize=(12,4))
plt.plot(np.arange(len(series)), series, label="History")
plt.plot(np.arange(len(series), len(series)+future_steps), preds, label="Forecast", color='red')
plt.legend()
plt.title("LSTM Future Forecasting (Next 100 steps)")
plt.show()

```
